Menggunakan HeatMap untuk mengetahui korelasi antar fitur. Korelasi yang memiliki nilai tinggi, akan digunakan untuk klasifikasi




```
import pandas as pd
import numpy as np
import seaborn as sns

X = data.iloc[:,0:20]  
y = data.iloc[:,-1]    
# mendapatkan korelasi antar kolomnya
corrmat = data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(10,10))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```
df = data.drop(['movement_reactions','movement_balance','power_shot_power','movement_sprint_speed'], axis = 1)
```

Korelasi antar fitur

```
datset = df.select_dtypes(include = ['float64', 'int64',]).iloc[:, :].corr()
plt.figure(figsize=(5, 5))
ax = sns.heatmap(datset, vmax=1, square=True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```
datset = df.select_dtypes(include = ['float64', 'int64',]).iloc[:, :].corr()
sns.heatmap(datset, annot = True, cmap = 'coolwarm')
plt.show()
```
